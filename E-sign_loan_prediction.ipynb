## Restructuring Dataset

```python
data.head()
```

```python
data['Months_employed'] = (data['months_employed'] + data['years_employed']*12)
```

```python
data.head()
```

```python
data['Personal_account_months'] = (data['personal_account_m'] + data['personal_account_y']*12)
```

```python
data.drop(['months_employed','years_employed', 'entry_id', 'personal_account_m', 'personal_account_y'], axis=1, inplace=True)
```

```python
data.head()
```

## Correlation Matrix

```python
data1 = data.drop(columns = ["e_signed"])
```

```python
data1.corrwith(data['e_signed']).plot.bar(figsize = (20,10), title = "Correlation with E Signed", rot = 45, grid = True)
```

```python
corr = data.corr()
plt.figure(figsize = (20,10))
sns.heatmap(corr,annot = True)
```

## Spliting the Model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data1, data["e_signed"], test_size = 0.2, random_state = 0)
```

## Scaling values


```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
X_train = scale.fit_transform(X_train)
X_test = scale.transform(X_test)
```

```python
X_test, X_train
```

# Model Training

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression 
```

```python
logi = LogisticRegression()
logi.fit(X_train,y_train)
y_pred = logi.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
```

```python
accu= accuracy_score(y_test,y_pred)

f1 = f1_score(y_test,y_pred)    

precision = precision_score(y_test,y_pred)

recall = recall_score(y_test,y_pred)
accu,f1,precision, recall
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross = cross_val_score(logi,X_train,y_train,cv = 10,scoring = "accuracy").mean()
```

```python
results = pd.DataFrame([["LogisticRegression",cross,accu,f1,precision,recall]], columns = ["Model","Cross_val_score","Accuracy","F1","Precision","Recall"])
results
```

### SVM model

```python
from sklearn.svm import SVC
```

```python
svm = SVC(kernel = "linear")
svm.fit(X_train,y_train)
y_pred = svm.predict(X_test)
```

```python
accu= accuracy_score(y_test,y_pred)

f1 = f1_score(y_test,y_pred)    

precision = precision_score(y_test,y_pred)

recall = recall_score(y_test,y_pred)
accu,f1,precision, recall
```

```python
confusion_matrix(y_test,y_pred)
```

```python
cross = cross_val_score(svm,X_train,y_train, cv= 10, scoring = "accuracy").mean()
```

```python
results1 = pd.DataFrame([["SVM Classifier",cross,accu,f1,precision,recall]], columns = ["Model","Cross_val_score","Accuracy","F1","Precision","Recall"])
results = pd.concat([results,results1], ignore_index=True)
results
```
